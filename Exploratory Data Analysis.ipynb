### Sum of Sales Price for Missing Values against Sum of Sales Price for Non Missing Values for the Features 
- Check if there is a Depency or relationship

```python
for feature in missing_features:
    data1=data.copy()
    
    
    data1[feature]=np.where(data1[feature].isnull(),1,0)
    data1.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

From the Observation above it shows that there is somewhat relationship between a missing value sales price and non missing value  sale price so

# Discrete and Continous Variables
- Discrete data involves round, concrete numbers that are determined by counting. 
- Continuous data involves complex numbers that are measured across a specific time interval.

Basically, you can use the unique function to get the unique value and latwer you jusat found the count
Or you can even use the nunique() function.

# Feature Scaling
### Information Leakage

- Lots of ideas on Standardization  and Normalization
- You can fit on the training data and you will use that that to transform the testing ( information leakage)
- Or you can fot _transform them differently

```python
feature_scale=[i for i in data.columns if data[i].dtype !='O' and i not in [ 'SalePrice']]
len(feature_scale)
```

```python
from sklearn.preprocessing import MinMaxScaler
MinMax=MinMaxScaler()

MinMax.fit_transform(data[feature_scale])
```
